## fuction

```python
def dataframec_ss( df ):
    # Create a SentimentIntensityAnalyzer object
    sentiments = SentimentIntensityAnalyzer()

    # using polarity_scores method of SentimentIntensityAnalyzer to create new columns saving pos, neg, neu scores
    # sentiments.polarity_scores create a dict containing pos, neg, neu, and compound scores
    # pos, neu, and neg scores are ratios for proportions of text that fall in each category
    df['comment_pos'] = [ sentiments.polarity_scores(i)['pos'] for i in df['comment_text'] ]
    df['comment_neg'] = [ sentiments.polarity_scores(i)['neg'] for i in df['comment_text'] ]
    df['comment_neu'] = [ sentiments.polarity_scores(i)['neu'] for i in df['comment_text'] ]
    df['comment_compound'] = [ sentiments.polarity_scores(i)['compound'] for i in df['comment_text'] ]
    df['comment_sentiment'] = [ classifier_s(sentiments.polarity_scores(i)['compound']) for i in df['comment_text'] ]
```

```python
def text_ss(text):
    sentiments = SentimentIntensityAnalyzer()
    ss = { 'pos': sentiments.polarity_scores(text)['pos'], 'neg': sentiments.polarity_scores(text)['neg'],
           'neu': sentiments.polarity_scores(text)['neu'], 'compound': sentiments.polarity_scores(text)['compound'],
           'sentiment': classifier_s(sentiments.polarity_scores(text)['compound'])}
    return ss
```

```python
def classifier_s( c_score ):
    if c_score >= 0.05:
        return "positive"
    elif c_score <= - 0.05:
        return "negative"
    else:
        return "neutral"
```

```python
def avgcommentcompound( url, df ):
    comments_bool = pd.DataFrame( df['WebVideoURL'] == url )
    comments = df[ comments_bool["WebVideoURL"] ]
    avgcomment = [ comments['comment_compound'].mean(), classifier_s(comments['comment_compound'].mean())]
    return avgcomment
```

```python
def dataframep_ss( df, df_comments ):
    # Create a SentimentIntensityAnalyzer object
    sentiments = SentimentIntensityAnalyzer()

    pos = []
    neg = []
    neu = []
    compound = []
    sentiment = []
    avgcommentc = []
    avgcomments = []
    for i in range( len(df) ):
        d = text_ss( df['post_text'][i] )
        pos.append( d['pos'] )
        neg.append( d['neg'] )
        neu.append( d['neu'] )
        compound.append( d['compound'] )
        sentiment.append( d['sentiment'] )
        avgcommentc.append( avgcommentcompound(df['WebVideoURL'][i], df_comments)[0] )
        avgcomments.append( avgcommentcompound(df['WebVideoURL'][i], df_comments)[1] )

    df['post_pos'] = pos
    df['post_neg'] = neg
    df['post_neu'] = neu
    df['post_compound'] = compound
    df['post_sentiment'] = sentiment
    df['post_avgcommentcompound'] = avgcommentc
    df['post_avgcommentsentiment'] = avgcomments
```

```python
def counter_cs( df ):
    value = df['comment_sentiment'].value_counts()
    l = [ value['positive'], value['negative'], value['neutral'] ]
    return l
```

### Sentiment Score
