```python
import pandas as pd
import string
from string import punctuation
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
import numpy as np
from PIL import Image
from collections import Counter

import nltk

from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
from nltk.stem.lancaster import LancasterStemmer
lancaster_stemmer = LancasterStemmer()
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer('english')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

from nltk.corpus import stopwords
stopword = stopwords.words('english')
stopword += stopwords.words('Arabic')
stopword += stopwords.words('Danish')
stopword += stopwords.words('Dutch')
stopword += stopwords.words('Finnish')
stopword += stopwords.words('French')
stopword += stopwords.words('German')
stopword += stopwords.words('Hungarian')
stopword += stopwords.words('Italian')
stopword += stopwords.words('Norwegian')
stopword += stopwords.words('Portuguese')
stopword += stopwords.words('Romanian')
stopword += stopwords.words('Russian')
stopword += stopwords.words('Spanish')
stopword += stopwords.words('Swedish')
```

### clean

```python
def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text = " ".join(text)
    text = [snowball_stemmer.stem(word) for word in text.split(' ')]
    text = " ".join(text)
    return text

df2['comment_text'] = df2['comment_text'].apply(clean)
df2
```

```python
def demoji(text):
    emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F" # emoticons
    u"\U0001F300-\U0001F5FF" # symbols & pictographs
    u"\U0001F680-\U0001F6FF" # transport & map symbols
    u"\U0001F1E0-\U0001F1FF" # flags (iOS)
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
    u"\U00010000-\U0010ffff"
    "]+", flags=re.UNICODE)
    return(emoji_pattern.sub(r'', text))

df2['comment_text'] = df2['comment_text'].apply(demoji)
df2
```

```python
text = " ".join(i for i in df2['comment_text'])
unicode_text = text.encode('utf-8').decode('utf-8')
word_list = text.split()
words = Counter(word_list)
words_most = Counter( words ).most_common(75)
mask = np.array(Image.open("tiktok.jpg"))
#plt.rcParams['font.family'] = 'Arial Unicode MS'
worddog = WordCloud( width = 1000, height =800, mask = mask, stopwords = stopword, scale = 10, background_color = "white", colormap = "Set1", max_words = 75, min_font_size = 8, margin = 3, collocations = False).generate(unicode_text)
plt.figure( figsize=(20,10) ) 
plt.imshow(worddog, interpolation='bilinear')
plt.axis("off")
```

```python
postext = " ".join( [i for i in df2['comment_text'][df2['comment_sentiment'] == 'positive']] )
posunicode_text = postext.encode('utf-8').decode('utf-8')
posword_list = postext.split()
poswords = Counter(posword_list)
poswords_most = Counter( poswords ).most_common(75)
mask = np.array(Image.open("tiktok.jpg"))
#plt.rcParams['font.family'] = 'Arial Unicode MS'
posworddog = WordCloud( width = 1000, height =800, mask = mask, stopwords = stopword, scale = 10, background_color = "white", colormap = "Set3", max_words = 75, min_font_size = 8, margin = 3, collocations = False).generate(posunicode_text)
plt.figure( figsize=(20,10) ) 
plt.imshow(posworddog, interpolation='bilinear')
plt.axis("off")
```

```python
negtext = " ".join( [i for i in df2['comment_text'][df2['comment_sentiment'] == 'negative']] )
negunicode_text = negtext.encode('utf-8').decode('utf-8')
negword_list = negtext.split()
negwords = Counter(negword_list)
negwords_most = Counter( negwords ).most_common(75)
mask = np.array(Image.open("tiktok.jpg"))
#plt.rcParams['font.family'] = 'Arial Unicode MS'
negworddog = WordCloud( width = 1000, height =800, mask = mask, stopwords = stopword, scale = 10, background_color = "white", colormap = "Dark2", max_words = 75, min_font_size = 8, margin = 3, collocations = False).generate(negunicode_text)
plt.figure( figsize=(20,10) ) 
plt.imshow(negworddog, interpolation='bilinear')
plt.axis("off")
```

```python
date_time = []
year = []
for time in df1['post_createTime']:
    dt = datetime.datetime.fromtimestamp( time )
    date_time.append( dt )
    year.append( dt.year )
    
df1.insert( 2, column = 'post_createDatetime', value = date_time )
df1.insert( 3, column = 'post_createYear', value = year )

df1 = df1.sort_values( by = ['post_createYear'] )
df1
```

```python
# separate df1 by year
group = df1.groupby( 'post_createYear' )
timedf = {} 

# save g_df(table "posts") and c_df(table "comments") as different csv(2020, 2021, 2022, 2023)
for y, g_df in group:
    print( y, 'table \"comments\"', sep = '' )
    df0 = g_df[ ['WebVideoURL'] ]
    c_df = df0.merge( df2, how = 'left' )
    display( c_df )
    timedf[y] = c_df
```

```python
# create function to count the total sentiment category in different year
def counter( df ):
    value = df['comment_sentiment'].value_counts()
    l = [ value['neutral'], value['positive'],value['negative'] ]
    return l

timetable = pd.DataFrame( {'2019': counter(timedf[2019]),
                           '2020': counter(timedf[2020]),
                           '2021': counter(timedf[2021]),
                           '2022': counter(timedf[2022]),
                           '2023': counter(timedf[2023]) }, index = ['neutral', 'positive', 'negative'] )
display( timetable )
```

```python
# create figure
fig = plt.figure( figsize = (180, 120) )
colors = ['#75A999', '#81ABC5', '#ADC992', '#FED872', '#FE9148']

# plot bar graph
timetable.plot( kind="bar", stacked = True, color = colors )
plt.title( "Sentiment of comments from 2019 to 2023" )
plt.xlabel( "Sentiment" )
plt.ylabel( "comments" )
plt.ylim( 0, 1800000 )

plt.savefig("barplot.png", bbox_inches='tight', facecolor = 'white' )
plt.close()
```
