### Split df by year

```python
date_time = []
year = []
for time in df1['post_createTime']:
    dt = datetime.datetime.fromtimestamp( time )
    date_time.append( dt )
    year.append( dt.year )
    
df1.insert( 2, column = 'post_createDatetime', value = date_time )
df1.insert( 3, column = 'post_createYear', value = year )

df1 = df1.sort_values( by = ['post_createYear'] )
df1
```
