```python
import pandas as pd
import string
from string import punctuation
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
import numpy as np
from PIL import Image
from collections import Counter

import nltk

from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
from nltk.stem.lancaster import LancasterStemmer
lancaster_stemmer = LancasterStemmer()
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer('english')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

from nltk.corpus import stopwords
stopword = stopwords.words('english')
stopword += stopwords.words('Arabic')
stopword += stopwords.words('Danish')
stopword += stopwords.words('Dutch')
stopword += stopwords.words('Finnish')
stopword += stopwords.words('French')
stopword += stopwords.words('German')
stopword += stopwords.words('Hungarian')
stopword += stopwords.words('Italian')
stopword += stopwords.words('Norwegian')
stopword += stopwords.words('Portuguese')
stopword += stopwords.words('Romanian')
stopword += stopwords.words('Russian')
stopword += stopwords.words('Spanish')
stopword += stopwords.words('Swedish')
```

# clean

```python
df = pd.read_csv('stayhomestaysafe_cleaned_comments.csv', encoding='utf-8')
def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text = " ".join(text)
    text = [snowball_stemmer.stem(word) for word in text.split(' ')]
    text = " ".join(text)
    return text

df['comment_text'] = df['comment_text'].apply(clean)
df
```

```python
def demoji(text):
    emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F" # emoticons
    u"\U0001F300-\U0001F5FF" # symbols & pictographs
    u"\U0001F680-\U0001F6FF" # transport & map symbols
    u"\U0001F1E0-\U0001F1FF" # flags (iOS)
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
    u"\U00010000-\U0010ffff"
    "]+", flags=re.UNICODE)
    return(emoji_pattern.sub(r'', text))

df['comment_text'] = df['comment_text'].apply(demoji)
df
```

# word cloud

```python
text = " ".join(i for i in df['comment_text'])
unicode_text = text.encode('utf-8').decode('utf-8')
#print(text)
word_list = text.split()
#print(word_list)
```

```python
words = Counter(word_list)
words_most = Counter( words ).most_common(75)
words_most
```

```python
mask = np.array(Image.open("tiktok.jpg"))
#plt.rcParams['font.family'] = 'Arial Unicode MS'
worddog = WordCloud(font_path='arialuni.ttf', width = 1000, height =800, mask = mask, stopwords = stopword, scale = 10, background_color = "white", colormap = "Set3", max_words = 75, min_font_size = 8, margin = 3, collocations = False).generate(unicode_text)
plt.figure( figsize=(20,10) )
plt.imshow(worddog, interpolation='bilinear')
plt.axis("off")
```
