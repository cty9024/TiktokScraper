## data pre-processing - data cleaning

### comments

```python
row, column = df4.shape
print( 'the total rows of table "comments":', row )
```

```python
# checking the number of duplicates in table "comments"
dup = df4.duplicated().sum()
print( 'the total duplicated rows of table "comments":', dup )
df4c = df4.drop_duplicates() # create the dataFrame(df4c) save the data that duplicates was cleaned
print( '-> drop', dup, 'rows from df4' )
```

```python
# checking the duplicates in table "comments" cleaned
print( df4c.duplicated().sum() )
```

```python
# checking the number of missing values in each column from table "comments"
mis_value = df4c.isnull().sum()
print( 'the total number of the missing values in each columns:\n', mis_value, sep='' )
```

```python
# create the Boolean dataFrame(mis_bool4) indicating whether values are missing in table "comments" or not
mis_bool4 = df4c.isnull( )

# create the dataFrame(mis_df4) including the row having missing values in df4c
colname_df4 = [ 'comment_createTime', 'comment_diggCount', 'WebVideoURL', 'comment_replyCommentTotal',
                'comment_userName' ]
mis_df4 = df4c[ mis_bool4['comment_text'] ]

for name in colname_df4: 
    mis_df4 = pd.concat( [mis_df4, df4c[mis_bool4[name]]], axis=0 ) # merge
    
mis_df4 = mis_df4.drop_duplicates() # delete duplicates rows
display( mis_df4 )
```

```python
# create the Boolean dataFrame(mis_bool4) indicating if values are missing in table "comments"
mis_bool4 = df4c.isnull( )

# According to columns in table "comments" -> show the rows having missing values in df4c
colname_df4 = [ 'comment_createTime', 'comment_diggCount', 'WebVideoURL', 'comment_replyCommentTotal',
                'comment_userName', 'comment_text' ]
for name in colname_df4:
    print( 'the missing data in df4c(table "comments") column -', name )
    display( df4c[ mis_bool4[name] ] )
    print( '\n' )
```

```python
# create the dictionary save the mis_df4 index as key and the total missing values of the index row as value
mis_rowvalue = {}
for i in range( len(mis_df4) ):
    index = mis_df4.index[i]
    mis_rowvalue[ str(index) ] = mis_df4.loc[index].isnull().sum()

# show the possible quantity of missing values in each row from mis_df4
mis_tvalue = list( set(mis_rowvalue.values()) )
print( mis_tvalue )
```

```python
for i in mis_tvalue:
    l = []
    print( 'total missing values of each row in df4:', i )
    for k, v in mis_rowvalue.items():
        if v == i:
            l.append( k )
    print( 'total rows:', len(l) )
    print( 'row:', l, '\n' )
```

```python
# remove the row having missing values in df4c
bc = len(df4c)
df4c.dropna( inplace = True )
print( '-> drop', bc-len(df4c), 'rows from df4c' )
```

```python
# checking that the missing value in df4 was cleaned
print( 'the total number of the missing values in each columns:\n', df4c.isnull().sum(), sep='' )
```

```python
# checking whether the table "comments" use the wrong "WebvideoURL" in df1 to download comments
wrong_links = list( set(df4c['WebVideoURL']) - set(df1['WebVideoURL']) )
wrong_links
```

```python
drop_index = [] # record the index of wrong data
for row in wrong_links:
    # create the dataFrame(mis_dft) including the wrong data in df4 "WebVideoURL" 
    wrong_bool = pd.DataFrame( df4c['WebVideoURL'] == row )
    wrong_dfw = df4c[ wrong_bool["WebVideoURL"] ] 
    display( wrong_dfw )
    for i in range( len(wrong_dfw) ):
        index = wrong_dfw.index[i]
        drop_index.append( index )

print( 'the total rows of wrong data:', len(drop_index) )
```

```python
# remove the missing "WebVideoURL" in table "comments" from table "post"
df4c.drop( labels = drop_index, axis = 0, inplace = True )
print( '-> drop', len(drop_index), 'rows from df' )
```

```python
# checking the number of missing values in each column from table "posts", "authorMeta", "musicMeta"(df)
mis_value = dfc.isnull().sum()
print( 'the total number of the missing values in each columns:\n', mis_value, sep='' )
```

```python
# create the Boolean dataFrame(mis_bool) indicating if values are missing in post_text and post_musicURL columns
df5 = dfc[ ["post_text", "post_musicURL_x"] ] # create the dataFrame(df5) that its columns including the missing values
mis_bool = df5.isnull( ) 
mis_bool
```

```python
# create the dataFrame(mis_dft) including the missing data in df "post_text" 
mis_dft = dfc[ mis_bool["post_text"] ] 
display( mis_dft )

# if you need to drop the missing values, you should use "dropl_index" to record their index
dropl_index = []
for i in range( len(mis_dft) ):
    index = mis_dft.index[i]
    dropl_index.append( index )
    print( 'the total missing values of row', index, ':', mis_dft.loc[index].isnull().sum() )
```

```python
# create the dataFrame(mis_dfm) including the missing data in df "post_musicURL" 
mis_dfm = dfc[ mis_bool["post_musicURL_x"] ] 
display( mis_dfm )

for i in range( len(mis_dfm) ):
    index = mis_dfm.index[i]
    print( 'the total missing values of row', index, ':', mis_dfm.loc[index].isnull().sum() )
```

```python
# remove rows by the corresponding index missing value in mis_dft
dfc = df.drop( labels = dropl_index, axis = 0 )
print( '-> drop', len(df)-len(dfc), 'rows from df' )
```

```python
# clean "comments" table again
drop2_index = [] 
for row in mis_dft['WebVideoURL']:
    # create the dataFrame(mis_dft) including the wrong data in df4 "WebVideoURL" 
    delete_bool = pd.DataFrame( df4c['WebVideoURL'] == row )
    delete_df4c = df4c[ delete_bool["WebVideoURL"] ] 
    display( delete_df4c )
    for i in range( len(delete_df4c) ):
        index = delete_df4c.index[i]
        drop2_index.append( index )

print( 'the total rows of data which need to be delete:', len(drop2_index) )
```

```python
# remove the missing "WebVideoURL" in table "comments" from table "post"
df4c.drop( labels = drop2_index, axis = 0, inplace = True )
print( '-> drop', len(drop2_index), 'rows from df4c' )
```

```python
# checking whether the table "comments" save the comments downloaded from each "WebvideoURL" in dfc
mis_keys = list( set(dfc['WebVideoURL']) - set(df4c['WebVideoURL']) )
mis_keys
```
